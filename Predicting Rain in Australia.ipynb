```python
#DATA PREPROCESSING
#The variables we will use in our model would be everything above except Date,location,RainToday
new_data = data.drop(['Date','Location','RainToday'],axis=1)
new_data.info()
```

```python
#Using a pairplot to see some correlations
sns.pairplot(new_data[:1000],hue='RainTomorrow')
```

```python
#Converting the yes no to 1 and 0, also removing rows which are NA
def convert(string):
    if string=='No':
        string=0
    elif string=='Yes':
        string = 1
    return string
new_data['RainTomorrow']=new_data['RainTomorrow'].apply(convert)
new_data.dropna(inplace=True)
new_data.info()
```

```python
#Converting wind gust and direction to numerical 
final_data=pd.get_dummies(new_data,columns=['WindGustDir','WindDir9am','WindDir3pm'])
final_data.info()
```

```python
#MODEL BUILDING
from sklearn.model_selection import train_test_split
X = final_data.drop('RainTomorrow',axis=1)
y = final_data['RainTomorrow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
#PREDICTIONS AND EVALUATION
predictions = logmodel.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```
